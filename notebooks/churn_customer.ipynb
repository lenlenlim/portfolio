## I. Dataset description

<h3><b>Định nghĩa Customer churn</b></h3>

Customer churn là thuật ngữ trong kinh doanh, được hiểu là sự rời đi của khách hàng. Customer churn rate có nghĩa là tỷ lệ khách hàng rời đi trong một khoảng thời gian nhất định.
Trong kinh doanh, nếu khách hàng cảm thấy không hài lòng, thỏa mãn với dịch vụ và sản phẩm mà bạn cung cấp, họ sẽ rời đi, hay nói cách khác, họ sẽ ngừng kết nối hoặc hợp tác với doanh nghiệp của bạn. 

<h3><b>Tầm quan trọng của Customer churn đối với doanh nghiệp</b></h3>

- ***Thứ nhất***, vấn đề về thâm hụt tài chính. Rõ ràng, số lượng khách hàng rời đi càng nhiều, doanh thu và lợi nhuận của doanh nghiệp sẽ giảm, thậm chí là giảm ở con số lớn. Bên cạnh đó, theo trang Forrester, chi phí để sở hữu được một khách hàng mới nhiều gấp 5 lần chi phí để giữ một khách hàng cũ. Đặc biệt, chi phí để sở hữu một khách hàng trung thành mới nhiều gấp 16 lần chi phí cho khách hàng trung thành hiện tại. Như vậy, nếu mất khách hàng hiện tại, doanh nghiệp của bạn vừa mất một khoản doanh thu khá lớn cũng như tốn nhiều chi phí để kêu gọi, thu hút khách hàng mới.

- ***Thứ hai***, doanh thu sẽ tăng nếu customer churn rate thấp. Doanh nghiệp càng giữ được nhiều khách hàng, lợi nhuận càng cao. Theo báo cáo của trường đại học Harvard, trung bình, cứ 5% khách hàng được giữ lại, thì lợi nhuận thu vào sẽ dao động từ 25% đến 95%, những con số mà chúng ta chưa từng nghĩ tới. 
Vì vậy, việc quan tâm, chú ý đến Churn customer rate là cấp thiết để duy trì và phát triển doanh nghiệp.



>Đối với dịch vụ **"Heo số"**, nhóm sẽ định nghĩa một khách hàng là **churn customer** nếu dữ liệu của họ thỏa mãn hai điều kiện sau:
>*   Thứ nhất, khách hàng đó đã sử dụng dịch vụ "Heo số" ~ trường  **```totalSavings2021_heoSo```** không bị <font color='red'>**NULL**</font>

> Sau khi đã xác định được churn customer, team sẽ sử dụng dữ liệu về **thông tin cá nhân của khách hàng** cùng với những **hành vi** của khách hàng được ghi nhận **trong tháng 3** (các trường có chứa <font color='red'>**mar2021**</font> hoặc <font color='red'>**Mar2021**</font>) để xây dựng mô hình dự đoán churn customer.<br>

* `saving_value_3`: Tổng số dư trong app tháng 3
* `saving_value_6`: Tổng số dư trong app tháng 6

<h3><b>Những điều kiện để một khách hàng được dự báo sẽ là <font color="red">churn customer</font> trong tương lai</b></h3>

* Thứ nhất, khách hàng đó phải là người đã hoặc đang sử dụng dịch vụ **Heo Số**. Bằng cách lựa chọn những khách hàng có dữ liệu về tổng số lượt login vào dịch vụ **Heo Số** (`total_login_heoso`) trong năm 2021 (not <font color='red'>**NULL**</font>) 



````python
df = df[df['total_login_heoso'].notnull()].reset_index()
````

* Thứ hai, dữ liệu của khách hàng đó không ghi nhận **số dư tiết kiệm** trong cả tháng 3 và tháng 6 (`saving_value_3` và `saving_value_6`)

Từ đó ta sẽ có thêm hai điều kiện `churn_cond_0` và `churn_cond_1` như sau:


````python
churn_cond_0 = (df["saving_value_3"] == 0) & (df["saving_value_6"] == 0)
churn_cond_1 = df["saving_value_3"].isnull() & df["saving_value_6"].isnull()
````

* Thứ ba, nếu như dữ liệu khách hàng đó không ghi nhận **lượt login** nào vào dịch vụ **Heo Số** trong cả tháng 3 và tháng 6 (`total_login_3` và `total_login_6`) thì khả năng cao khách hàng đó cũng sẽ là churn customer

Điều kiện `churn_cond_2` sẽ thể hiện điều này:


````python
churn_cond_2 = (df['total_login_3'] == 0) & (df['total_login_6'] == 0)
````

 Nhóm cũng đã kiểm tra các điều kiện khác "tự nhiên hơn" để xác định **churn customer**, tuy nhiên những trường hợp đó lại không xảy ra với bộ data được cung cấp:

 >**Ghi nhận** số dư tiền gửi vào tháng 3 nhưng **không ghi nhận** số dư tiền gửi vào tháng 6:

````python
# Ghi nhận số dư vào tháng 3 nhưng không ghi nhận số dư vào tháng 6:
count_cond = (df["saving_value_3"].notnull() & df["saving_value_6"].isnull()).sum()

print(f'Số khách hàng ghi nhận số dư vào tháng 3 nhưng không ghi nhận số dư vào tháng 6: {count_cond}')
````

Kết hợp cả ba điều kiện `churn_cond_0`, `churn_cond_1` và `churn_cound_2`, ta sẽ tạo được label `is_churn`:

````python
df['is_churn'] = (churn_cond_0 | churn_cond_1 | churn_cond_2).astype('int')
````

- is_churn = 0: Khách hàng vẫn tiếp tục sử dụng Heo số
- is_churn = 1: Khách hàng rời bỏ

### Handling outliers

````python
import numpy as np

def Zscore_outlier(col):
    out = {col: 0}
    m = np.mean(df[col])
    sd = np.std(df[col])
    for i in df[col]: 
        z = (i - m) / sd
        if np.abs(z) > 3.0: # Nhóm định nghĩa Outliers có z-score > 3.0
            out[col] += 1
    return out

# Numeric features
numeric_columns = df.select_dtypes(include=['number']).drop(['service_code_week', 'service_code_month', 'is_churn'], axis=1)

print('Số lượng outliers (điểm dữ liệu có z-score > 3.0) là:')
for col in numeric_columns:
    print(f'{col} : {Zscore_outlier(col)[col]}')
````

````python
# Xóa các dòng chứa outlier
for col in numeric_columns:
    z_outliers = np.abs(stats.zscore(df[col])) > 3.0
    df = df[~z_outliers]

print("Số dòng còn lại sau khi loại bỏ outlier:", len(df))
````

## VI. Modeling

Dưới đây sẽ là hàm dùng để tính các metrics -> chọn ra những classifiers tốt nhất cho bộ dữ liệu này:

- Hàm này sẽ tiến hành chia tập train thành các folds, sau đó tiến hành resampling cho tập train

- Trong mỗi iteration, ta cũng tiến hành luôn đồng thời áp dụng scaler vào bộ dữ liệu (`fit_transform` trên tập train, `transform` trên tập evaluate)

- Kết quả trả về là kết quả: accuracy, precision, recall và f1_score trung bình của tập train và tập evaluate

````python
#@title    
def score_model(model, scaler, resolve_imblance, X_train, y_train):
    """
    Creates folds manually, and upsamples within each fold.
    Returns an array of validation (precision) scores
    """

    cv = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)
    imblance_resolver = resolve_imblance(random_state=42)
    columns_to_scale =  df.select_dtypes('number').drop(['is_churn','total_act_6', 'total_amt_6','saving_value_6','total_login_6',
                                                         'most_act_6_Giao Dich Phi Tai Chinh',
             'most_act_6_Giao Dich Rut Tien Atm', 'most_act_6_Giao Dich Thanh Toan', 'most_act_6_Giao Dich Tiet Kiem'], axis=1).columns
    scaler = scaler()
    scores = []

    X_train.reset_index(drop=True, inplace=True)  # Reset indices of X_train
    y_train.reset_index(drop=True, inplace=True)  # Reset indices of y_train

    num_fold = 0

    for train_fold_index, val_fold_index in cv.split(X_train, y_train):
        num_fold += 1
        # Get the training data
        X_train_fold, y_train_fold = X_train.iloc[train_fold_index], y_train[train_fold_index]

        # Get the validation data
        X_val_fold, y_val_fold = X_train.iloc[val_fold_index], y_train[val_fold_index]

        # Robust scale
        X_train_fold_scaled = scaler.fit_transform(X_train_fold[columns_to_scale])
        X_val_fold_scaled = scaler.transform(X_val_fold[columns_to_scale])

        # Resample only the data in the training section
        X_train_fold_scaled_upsample, y_train_fold_upsample = \
             imblance_resolver.fit_resample(X_train_fold_scaled,y_train_fold)

        # Fit the model on the resampled training data
        model_obj = model().fit(X_train_fold_scaled_upsample, y_train_fold_upsample)
        
        # Prediction
        y_pred_train_fold = model_obj.predict(X_train_fold_scaled_upsample)
        y_pred_val_fold = model_obj.predict(X_val_fold_scaled)

        # Score the metrics on the (resampled) traing data
        acc_train = accuracy_score(y_train_fold_upsample, y_pred_train_fold)
        precision_train = precision_score(y_train_fold_upsample, y_pred_train_fold)
        recall_train = recall_score(y_train_fold_upsample, y_pred_train_fold)
        f1_train = f1_score(y_train_fold_upsample, y_pred_train_fold)

        # Score the metrics on the (non-resampled) validation data
        acc_val = accuracy_score(y_val_fold, y_pred_val_fold)
        precision_val = precision_score(y_val_fold, y_pred_val_fold)
        recall_val = recall_score(y_val_fold, y_pred_val_fold)
        f1_val = f1_score(y_val_fold, y_pred_val_fold)

        scores.append(
              [acc_train,acc_val,precision_train,precision_val, recall_train,recall_val, f1_train,f1_val]
            )

    # Caculate the mean of these folds
    acc_train = np.mean([scores[i][0] for i in range(0,5)])
    acc_val = np.mean([scores[i][1] for i in range(0,5)])
    precision_train = np.mean([scores[i][2] for i in range(0,5)])
    precision_val = np.mean([scores[i][3] for i in range(0,5)])
    recall_train = np.mean([scores[i][4] for i in range(0,5)])
    recall_val = np.mean([scores[i][5] for i in range(0,5)])
    f1_train = np.mean([scores[i][6] for i in range(0,5)])
    f1_val = np.mean([scores[i][7] for i in range(0,5)])

    return acc_train, acc_val, precision_train,precision_val, recall_train, recall_val, f1_train,f1_val
````

Ta sẽ thử nghiệm với một số classifier phổ biến:

#### Sử dụng Up Sampling (SMOTE)

````python
metrics_result_SMOTE = {} 
for classifier in classifiers.keys(): 
  time_start = time.time()
  print(f'Prediting by classifier: {classifiers[classifier]} ....')
  acc_train, acc_val, pre_train, pre_val, re_train, re_val, f1_train, f1_val = \
      score_model(classifier, RobustScaler, SMOTE, X_train, y_train)
  print(f'Time consumed: {time.time() - time_start}')
  if classifiers[classifier] not in metrics_result_SMOTE:
    metrics_result_SMOTE[classifiers[classifier]] = [acc_train, acc_val, pre_train, pre_val, re_train, re_val, f1_train, f1_val]
    print('-' * 20)
````

````python
model_rf = RandomForestClassifier(**grid_search_rf.best_params_)
model_xgb = XGBClassifier(**grid_search_xgb.best_params_)

def plot_learning_curve(estimator, X_train, y_train, train_sizes=np.linspace(.1, 1.0, 5), ax=None, cv=None, estimator_name=None):
    train_sizes, train_scores, test_scores = learning_curve(
        estimator, X_train, y_train, cv=cv, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    ax.fill_between(train_sizes, train_scores_mean - train_scores_std,
                     train_scores_mean + train_scores_std, alpha=0.1,
                     color="#ff9124")
    ax.fill_between(train_sizes, test_scores_mean - test_scores_std,
                     test_scores_mean + test_scores_std, alpha=0.1, color="#2492ff")
    ax.plot(train_sizes, train_scores_mean, 'o-', color="#ff9124",
             label="Training score")
    ax.plot(train_sizes, test_scores_mean, 'o-', color="#2492ff",
             label="Cross-validation score")
    ax.set_title(estimator_name, fontsize=14)
    ax.set_xlabel('Training size (m)')
    ax.set_ylabel('Score')
    ax.grid(True)
    ax.legend(loc="best")
````

Phân phối của các giá trị được dự đoán trên dữ liệu huấn luyện/kiểm tra

````python
def distribution_plot(ax, x_value, y_value, x_label, y_label, title):
    """
    Params:
     :x_value(array): 
     :y_value(array):
     :x_label(str): 
     :y_label(str):
     :title(str):
    
    Returns
    """
    sns.kdeplot(x_value, color="r", label=x_label, ax=ax)
    sns.kdeplot(y_value, color="b", label=y_label, ax=ax)

    ax.set_title(title)
    ax.legend()
````
